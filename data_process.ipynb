## Read data

```python
# sắp xếp tên thư mục giống trên window
import re
def sorted_alphanumeric(data):
    convert = lambda text: int(text) if text.isdigit() else text.lower()
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(data, key=alphanum_key)
```

```python
path = "/content/drive/MyDrive/Project/dataset/raw_data/dendo"
os.chdir(path)
sort_listdir = sorted_alphanumeric(os.listdir())
print(sort_listdir)
data_hanam = []
data_thainguyen = []
for i in sort_listdir:
    with open(path+"/"+i+"/info.txt", 'rb') as f:
      clean_lines = (line.replace(b',,',b',') for line in f)
      content = np.genfromtxt(clean_lines, delimiter=',')
      if i[:5] == "hanam":
            data_hanam.append(content)
      elif i[:10] == "thainguyen":
            data_thainguyen.append(content)
print("Ha Nam")
data_hanam[0]
```

## Lọc data

```python
def linear_equation(x1,y1,x2,y2):
    a = (y1-y2)/(x1-x2)
    c = y1-a*x1
    b = -1
    return a, b, c
```

```python
def safe(value1, value2):
    if value1*value2>=0:
        return True
    return False
```

Lọc data ở Hà Nam

```python
# Các điểm cho trước trên vùng lọc
x1 = 2.9411764705882035
y1 = 1142.8823529411764
x2 = 935.2941176470588
y2 = 701.7058823529411
x3 = 1752.941176470588
y3 = 725.235294117647
x4 = 2400.0
y4 = 1916.4117647058822
# Điểm để xét trong vùng đang quan sát
X = 1280
Y = 1920

# Đường thẳng lọc data hà nam
a1, b1, c1 = linear_equation(x1,y1,x2,y2)
a2, b2, c2 = linear_equation(x2,y2,x3,y3)
a3, b3, c3 = linear_equation(x3,y3,x4,y4)
```

```python
filtered_data_hanam = []

for datas in data_hanam:
    filters = []
    for data in datas:
        if (safe(a1*X+b1*Y+c1, a1*((data[4]+data[3])/2)+b1*((data[6]+data[5])/2)+c1)==True 
        and safe(a2*X+b2*Y+c2, a2*((data[4]+data[3])/2)+b2*((data[6]+data[5])/2)+c2)==True
        and safe(a3*X+b3*Y+c3, a3*((data[4]+data[3])/2)+b3*((data[6]+data[5])/2)+c3)==True):
            filters.append(data)
    filtered_data_hanam.append(filters)
```

```python
#lọc không vượt đèn đỏ
# Các điểm cho trước trên vùng lọc
x5 = 3.0
y5 = 1436.0
x6 = 645.0
y6 = 760.0
x7 = 1369.0
y7 = 762.0
x8 = 1096
y8 = 1919
# Điểm để xét trong vùng đang quan sát
X1 = 849.0
Y1 = 1070.0

# Đường thẳng lọc data thái nguyên không vượt đèn đỏ
a4, b4, c4 = linear_equation(x5,y5,x6,y6)
a5, b5, c5 = linear_equation(x6,y6,x7,y7)
a6, b6, c6 = linear_equation(x7,y7,x8,y8)

#lọc vượt đèn đỏ
#các điểm cho trước trên vùng lọc
x9 = 1277.0
y9 = 1918.0
x10 = 1469.0
y10 = 802.0
x11 = 1717.0
y11 = 386.0
x12 = 2559
y12 = 104
# Điểm để xét trong vùng đang quan sát
X = 1717.0
Y = 802.0

# Đường thẳng lọc data thái nguyên vượt đèn đỏ
a7, b7, c7 = linear_equation(x9,y9,x10,y10)
a8, b8, c8 = linear_equation(x10,y10,x11,y11)
a9, b9, c9 = linear_equation(x11,y11,x12,y12)
```

```python
filtered_data_thainguyen_kovuot = []
filtered_data_thainguyen_vuot = []
for datas in data_thainguyen:
    filters_kovuot = []
    filters_vuot = []
    for data in datas:
        if (safe(a4*X1+b4*Y1+c4, a4*((data[4]+data[3])/2)+b4*((data[6]+data[5])/2)+c4)==True 
        and safe(a5*X1+b5*Y1+c5, a5*((data[4]+data[3])/2)+b5*((data[6]+data[5])/2)+c5)==True
        and safe(a6*X1+b6*Y1+c6, a6*((data[4]+data[3])/2)+b6*((data[6]+data[5])/2)+c6)==True):
            filters_kovuot.append(data)
        if (safe(a7*X+b7*Y+c7, a7*((data[4]+data[3])/2)+b7*((data[6]+data[5])/2)+c7)==True 
        and safe(a8*X+b8*Y+c8, a8*((data[4]+data[3])/2)+b8*((data[6]+data[5])/2)+c8)==True
        and safe(a9*X+b9*Y+c9, a9*((data[4]+data[3])/2)+b9*((data[6]+data[5])/2)+c9)==True):
            filters_vuot.append(data)
    filtered_data_thainguyen_kovuot.append(filters_kovuot)
    filtered_data_thainguyen_vuot.append(filters_vuot)
```
